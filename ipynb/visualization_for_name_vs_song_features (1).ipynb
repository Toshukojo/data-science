```
small_df = small_df[small_df['year']>1986]
small_df_by_dance = small_df.sort_values('danceability', ascending=False)
small_df_by_dance.head()
```

```
for i in range(5):
    ZZZ = i
    row = small_df_by_dance.iloc[ZZZ]
    year = row[12]
    artists = row[13]
    song = f"#{ZZZ+1} Danceability --- Song Title : {row[0]}  --- Artist : {artists}  --- Year {year}"

    acous = row[1]
    dance = row[2]
    energ = row[3]
    instr = row[4]
    key = row[5]/10
    liven = row[6]
    loudn = row[7]/100 #100
    popul = row[8]/100 #100
    speec = row[9]
    tempo = row[10]/250 #max 250
    valen = row[11]
    # trimmed to article on spotify
    y = [acous, dance, energ,  liven,   speec, valen]
    x = ['acousticness', 'danceability', 'energy',   
        'liveness',  'speechiness',  'valence']

    # name vs song features
    df = small_df_by_name
    fig = px.bar(df, x=x, y=y, labels=x, title=song)
    fig.update_traces(marker_color='orange')
    # fig['layout']['yaxis'].update(range=[-1, 1], autorange=False)
    fig.show()
```
