```python
import pandas as pd

df = pd.read_csv("https://github.com/spotify-song-suggestion/data-science/raw/main/csv/data.csv")
```

```python
df.head()
```

```python
import datetime

def processDate(date_str):
    try:
        date = date_str.split("-")
        if len(date) == 1:
            return datetime.datetime(year=int(date[0]), month=1, day=1)
        elif len(date) == 2:
            return datetime.datetime(year=int(date[0]), month=int(date[1]), day=1)
        else:
            return datetime.datetime(year=int(date[0]), month=int(date[1]), day=int(date[2]))
    except Exception as e:
        print(e)
        print(date_str)
df["year"] = df["release_date"].apply(processDate)
```

```python
from sklearn.neighbors import NearestNeighbors
from sklearn.preprocessing import StandardScaler

columns = ["acousticness", "danceability", "energy", "instrumentalness", "liveness", "loudness", "speechiness", "tempo", "valence"]
properties_df = df[columns]
```

```python
scaled = StandardScaler().fit_transform(properties_df)
nn = NearestNeighbors(n_neighbors=10, n_jobs=-1).fit(scaled)
```

```python
distances, indices = nn.kneighbors(scaled[0:2])
```
